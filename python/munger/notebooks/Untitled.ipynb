```python
import umap
# import hdbscan
import numpy as numpy
import hdbscan
import pandas as pd
from PIL import Image
import sklearn.cluster as cluster
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
```

```python
def cluster(d): 
    d.thumbnail(size=(128, 128))
    d_prime = numpy.asarray(d).reshape(-1,3)
    clusterable_embedding = umap.UMAP(densmap=True, random_state=42, n_components=2).fit_transform(d_prime)
    hdbs = hdbscan.HDBSCAN(
        min_samples=10,
        min_cluster_size=100,
        )
    labels = hdbs.fit_predict(clusterable_embedding)
    probs = hdbs.probabilities_
    data = numpy.vstack([labels, numpy.apply_along_axis(str, 1, d_prime), probs ]).T
    df = pd.DataFrame(data, columns= ["label", "colors", "probs"])
    return df.sort_values('probs').drop_duplicates(['label'],keep='last')[["colors"]]
    
```
